# Live Updates: Spatial Transcriptomics in Pancreatic Cancer

This notebook queries PubMed for publications related to **spatial transcriptomics** in **pancreatic cancer** using title and abstract search, and generates a trend plot of publication counts per year.

```python
from Bio import Entrez
import datetime
import matplotlib.pyplot as plt
import os

# Set your email and (optionally) API key for NCBI
Entrez.email = 'your.email@example.com'
# Entrez.api_key = 'your_api_key'

# Define the search query
search_query = '"spatial transcriptomics"[Title/Abstract] AND "pancreatic cancer"[Title/Abstract]'

# Define the range of years (start at 2015)
start_year = 2015
current_year = datetime.datetime.now().year
years = list(range(start_year, current_year + 1))

publication_counts = []

for year in years:
    # Create a query with publication date limits for each year
    query = f'({search_query}) AND ("{year}"[Date - Publication] : "{year}"[Date - Publication])'
    handle = Entrez.esearch(db='pubmed', term=query)
    record = Entrez.read(handle)
    count = int(record.get('Count', 0))
    publication_counts.append(count)
    handle.close()

# Plot the publication trend
plt.figure(figsize=(10, 5))
plt.plot(years, publication_counts, marker='o')
plt.title('Publication Trend: Spatial Transcriptomics in Pancreatic Cancer (2015 to Present)')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.grid(True)

# Force x-axis to show every year
plt.xticks(years, [str(y) for y in years], rotation=45)  # rotation=45 if you want angled labels

# Create the 'plots' directory if it doesn't exist
if not os.path.exists('plots'):
    os.makedirs('plots')

# Save and display the plot
plt.savefig('plots/publication_trend.png', bbox_inches='tight')  # bbox_inches='tight' helps fit labels
plt.show()
```
